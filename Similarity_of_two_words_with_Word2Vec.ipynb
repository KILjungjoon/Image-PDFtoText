```python
# loaing a model
from gensim.models import word2vec
model=word2vec.Word2Vec.load("file_path/sg_15win_300vec_2min.model")
```

## get similarity of two words

```python
print(model.wv.similarity('눈01/NNG', '속01/NNG'))
print(model.wv.similarity('눈01/NNG', '안01/NNG'))
```

```python
import pandas as pd
import re
from gensim.models import Word2Vec

# Extract words with the highest similarity to the search term
search_term = '비01/NNG'
data = model.wv.most_similar(search_term, topn=100)

# save result to dataframe
result_data = {'word': [], 'information': [], 'similarity': []}
for index, value in enumerate(data):
    morph=re.sub(r'[^ㄱ-ㅎ가-힣]', '', value[0])
    info=re.sub(r'[ㄱ-ㅎ가-힣]', '', value[0])
    result_data['word'].append(morph)
    result_data['information'].append(info)
    result_data['similarity'].append(value[1])

result_df = pd.DataFrame(result_data)

# Output data frame
pd.set_option('display.max_rows', None)
result_df
```
